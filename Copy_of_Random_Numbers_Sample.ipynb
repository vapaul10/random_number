```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
## Generate 100 random Data Points along 3 dimensions
x, y, scale = np.random.randn(3, 150)
fig, ax = plt.subplots()

## Map each onto a scatterplot we'll create with Matplotlib
ax.scatter(x=x, y=y, c=scale, s=np.abs(scale)*500)
ax.set(title="Some random data, created with JupyterLab!")
plt.show()
```

```python
# prompt: Select top 5 random integers from above data points

x, y, scale = np.random.randn(3, 150)
top_5 = np.argsort(scale)[-5:]
fig, ax = plt.subplots()
ax.scatter(x=x[top_5], y=y[top_5], c=scale[top_5], s=np.abs(scale[top_5])*500)
ax.set(title="Some of the top 5 random data, created with JupyterLab!")
plt.show()
```
